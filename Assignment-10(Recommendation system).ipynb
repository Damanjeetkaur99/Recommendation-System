```python
import pandas as pd
import numpy as np
import chardet
```

```python
book=pd.read_csv("bookr.csv",encoding='Latin1')
df=pd.DataFrame(book)
df
```

```python
df1=df.iloc[:,1:]
df1
```

```python
df2=df1.rename(columns={'User.ID':'user_id','Book.Title':'book','Book.Rating':'book_rating'})
```

```python
df2.head()
```

### number of unique users in the dataset

```python
len(df2.user_id.unique())
```

```python
len(df2.book.unique())
```

```python
df2.isna().sum()
```

### converting data into wide format from long format

```python
user_book_df = df2.pivot_table(index='user_id',
                                 columns='book',
                                 values='book_rating',aggfunc='sum')
user_book_df
```

```python
user_book_df.index = df2.user_id.unique() #saving user id into index
user_book_df
```

### Imputing NaNs values with 0 values

```python
user_book_df.fillna(0, inplace=True)
```

```python
user_book_df
```

## Cosine Similarity between Users

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_book_df.values,metric='cosine')
user_sim
```

### converting it into dataframe

```python
user_sim_df = pd.DataFrame(user_sim)
```

## setting index and columns names to user_id

```python
user_sim_df.index = df2.user_id.unique()
user_sim_df.columns = df2.user_id.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

#### filling diagonal =0

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

### Most Similar 5 Users

```python
user_sim_df.idxmax(axis=1)[0:5]
```
